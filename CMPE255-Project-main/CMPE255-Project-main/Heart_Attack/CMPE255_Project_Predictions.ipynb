# Predictions

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
import numpy as np
```

```python
# Predict no heart attack or heart attack 
```

```python
# Get dataset
full_dataset = pd.read_csv('data/2015.csv')
preprocessed_dataset = pd.read_csv('data/preprocessed_dataset.csv')
```

```python
# Get rows where CVDINFR4 = 7 or 9 from the dataset to make predictions
dataset = full_dataset.loc[full_dataset['CVDINFR4'].isin([7, 9])]
dataset.reset_index(drop=True, inplace=True)
```

```python
dataset.head()
```

```python
# Get only the columns important for the model
prediction_dataset = dataset[preprocessed_dataset.columns]
```

```python
prediction_dataset.head()
```

```python
# Remove the CVDINFR4 column from the prediction dataset
prediction_dataset.drop(['CVDINFR4'], axis=1, inplace=True)
```

```python
# Normalize prediction dataset
scaler = StandardScaler()
prediction_dataset_norm = scaler.fit_transform(prediction_dataset)
```

```python
prediction_dataset_norm
```

```python
prediction_dataset_norm = pd.DataFrame(data = prediction_dataset_norm, columns=prediction_dataset.columns)
prediction_dataset_norm.head()
```

```python
# Get labels from preprocessed database
labels = preprocessed_dataset['CVDINFR4']
```

```python
labels
```

```python
# Remove the CVDINFR4 column from the preprocessing dataset
preprocessed_dataset.drop(['CVDINFR4'], axis=1, inplace=True)
```

```python
# Normalize preprocessed dataset
scaler = StandardScaler()
preprocessed_dataset_norm = scaler.fit_transform(preprocessed_dataset)
```

```python
preprocessed_dataset_norm = pd.DataFrame(data = preprocessed_dataset_norm, columns=preprocessed_dataset.columns)
preprocessed_dataset_norm.head()
```

```python
# Use XGBoost to make predictions
model = XGBClassifier(importance_type='gain', learning_rate=0.01, max_depth=5,  n_estimators=150, scale_pos_weight=33, 
                      use_label_encoder=False)
model.fit(preprocessed_dataset_norm, labels)
```

```python
# Predictions
y_pred = model.predict(prediction_dataset_norm)
```

```python
# Number of people predicted to be susceptible to Heart attack
np.count_nonzero(y_pred==1)
```

```python
# Number of people predicted to not have Heart attack
np.count_nonzero(y_pred==0)
```
