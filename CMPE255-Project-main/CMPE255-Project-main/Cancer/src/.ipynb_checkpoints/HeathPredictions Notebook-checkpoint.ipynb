```python
nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
  
df.dropna(how='all', axis=1, inplace=True)

df
```

```python
print("count of NULL values before imputation\n")
df.isnull().sum()
```

```python
df = df.replace(np.nan, 0)
```

```python
print("count of NULL values after imputation\n")
df.isna().sum()
```

```python
df = df.drop(['IDATE', 'IMONTH', 'IDAY', 'IYEAR', 'EXACTOT1', 'EXACTOT2', 'PCDMDECN'], axis=1)
```

```python
display(df.head())
```

```python

X = df.iloc[:,:-1]
X
```

```python

Y = df.iloc[:,-1]
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, 
                                                    train_size=0.67, 
                                                    random_state=42,
                                                    stratify=Y)

print(f"Train labels:\n{y_train}")
print(f"Test labels:\n{y_test}")
```

```python
numeric_cols = X_train.select_dtypes(include=['float64','float32']).columns
numeric_cols.shape

set(X_train.columns.tolist())-set(numeric_cols.tolist())
```

```python
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
plot_confusion_matrix(classifier, X_test, y_test, display_labels=['alt.atheism',
 'comp.graphics'], cmap=plt.cm.Blues)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```
